```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
X = np.array([[5,3],[10,15],[24,10],[30,45],[40,20],[55,70],[71,80],[60,78],[90,91],[50,30],[15,45],[20,60],[60,20],[10,24],[70,60]])
labels = range(1,16)
plt.figure(figsize=(10,7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0], X[:,1])
for label, x, y in zip(labels, X[:,0], X[:,1]):
  plt.annotate(label, xy= (x,y), xytext=(-3,3), textcoords = 'offset points', ha='right', va='bottom')
plt.show()
```

```python
#grafica dendograma
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(X, 'single')
labellist = range(1,16)
plt.figure(figsize=(10,7))
dendrogram(linked, orientation = 'top', labels = labellist, distance_sort='descending', show_leaf_counts=True)
plt.show()
```

```python
#aplicamos el algoritmo jerarquico
from sklearn.cluster import  AgglomerativeClustering
cobweb = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cobweb.fit_predict(X)
labels = cobweb.labels_
plt.scatter(X[:, 0], X[:,1], c=labels,cmap='rainbow')
plt.show()
```
