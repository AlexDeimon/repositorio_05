```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets
#abrir el dataset boston
boston = datasets.load_boston()
print(boston.keys())
print(boston.data.shape)
```

```python
#seleccion numero de cuatos y valor vivienda
X = boston.data[:, np.newaxis, 5]
y = boston.target
```

```python
#graficar
plt.figure(figsize=(10,7))
plt.scatter(X, y, color  = 'blue')
plt.title('Date Boston Housing')
plt.xlabel('numero de habitaciones')
plt.ylabel('valor promedio')
plt.show()
```

```python
#implementamos algoritmo de regresion lineal
#entrenamiento y test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.2)
lr = linear_model.LinearRegression()
lr.fit(X_train,Y_train)
Y_pred = lr.predict(X_test)
```

```python
#graficar
plt.figure(figsize=(10,7))
plt.scatter(X_test, Y_test, color  = 'blue')
plt.plot(X_test,Y_pred, color = 'red', linewidth = 5)
plt.title('Modelo regresion lineal Date Boston')
plt.xlabel('numero de habitaciones')
plt.ylabel('valor promedio')
plt.show()
```

```python
print('Datos del modelo de regresion lineal')
print('Valor de la pendiente / coeficiente a = ',lr.coef_)
print('valor del intersecto / coeficiente b = ', lr.intercept_)
```

```python
precision = lr.score(X_train,Y_train)
print(precision)
```
