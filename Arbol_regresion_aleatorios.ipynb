```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
data = np.random.RandomState(1)
x = np.sort(5*data.rand(80,1), axis=0)
x
```

```python
y = np.sin(x).ravel()
y[::5]+=3*(0.5 - data.rand(16))
y
```

```python
#modelo de regresion
regressor_1 = DecisionTreeRegressor(max_depth=2)
regressor_2 = DecisionTreeRegressor(max_depth=5)
regressor_1.fit(x,y)
regressor_2.fit(x,y)
```

```python
x_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
x_test
```

```python
y_pred1 = regressor_1.predict(x_test)
y_pred2 = regressor_2.predict(x_test)
```

```python
plt.scatter(x, y, label = 'datos', c = 'orange')
plt.title('Arbol de decision')
plt.ylabel('Target')
plt.xlabel('Data')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, label = 'datos', c = 'orange')
plt.plot(x_test, y_pred1, color = 'cornflowerblue', label = 'max_depth = 2', linewidth = 2)
plt.title('Arbol de decision')
plt.ylabel('Target')
plt.xlabel('Data')
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, label = 'datos', c = 'orange')
plt.plot(x_test, y_pred2, color = 'yellowgreen', label = 'max_depth = 5', linewidth = 2)
plt.title('Arbol de decision')
plt.ylabel('Target')
plt.xlabel('Data')
plt.legend()
plt.show()
```
