```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv('Mall_Customers.csv')
print(data.shape)
print(data.head())
X = data.iloc[:, 3:5].values #ingreso anual vs puntaje de gasto
plt.scatter(X[:,0], X[:,1])
```

```python
#dendograma ward
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title('Dendogram customers')
dend = shc.dendrogram(shc.linkage(X, method='ward'))
```

```python
#dendograma average = distancia promedio
plt.figure(figsize=(10,7))
plt.title('Dendogram customers')
dend = shc.dendrogram(shc.linkage(X, method='average'))
```

```python
#dendograma single = distancia mas corta
plt.figure(figsize=(10,7))
plt.title('Dendogram customers')
dend = shc.dendrogram(shc.linkage(X, method='single'))
```

```python
#dendograma complete = distancia mas larga
plt.figure(figsize=(10,7))
plt.title('Dendogram customers')
dend = shc.dendrogram(shc.linkage(X, method='complete'))
```

```python
#grafica dendograma otro metodo
from scipy.cluster.hierarchy import dendrogram, linkage
linked = linkage(X, 'ward')
labellist = range(1,201)
plt.figure(figsize=(10,7))
dendrogram(linked, orientation = 'top', labels = labellist, distance_sort='descending', show_leaf_counts=True)
plt.show()
```

```python
#aplicamos el algoritmo jerarquico con 5 clusters
from sklearn.cluster import  AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5)
cluster.fit_predict(X)
labels = cluster.labels_
colores = ['blue','red','green','orange','magenta']
colores_cluster = [colores[labels[i]] for i in range(200)]
plt.figure(figsize=(10,7))
plt.scatter(X[:,0],X[:,1], c = colores_cluster, marker='o', alpha=0.5)
```

```python
#aplicamos el algoritmo jerarquico con 7 clusters
cluster = AgglomerativeClustering(n_clusters=7)
cluster.fit_predict(X)
labels = cluster.labels_
plt.figure(figsize=(10,7))
plt.scatter(X[:, 0], X[:,1], c=labels,cmap='rainbow')
```
