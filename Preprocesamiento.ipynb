```python
#1.	Realice la limpieza y adecuaci√≥n de la data (ACTOR_VIAL), para que quede lista, con los siguientes datos:
import pandas as pd
df = pd.read_excel("Siniestros_Viales_Consolidados_Bogota_DC.xlsx",sheet_name = "ACTOR_VIAL")
df.head(10)
```

```python
#a.	Eliminar columnas como CODIGO_ACCIDENTE, CODIGO_ACCIDENTADO y VEHICULO
dfs = df.drop(['CODIGO_ACCIDENTE', 'CODIGO_ACCIDENTADO', 'VEHICULO'], axis=1)
dfs.head(10)
```

```python
#b.	Filtrar el a√±o 2020
df_2020 = dfs.copy()
df_2020["FECHA"] = pd.to_datetime(dfs['FECHA'])
df_2020 = df_2020.query("FECHA >= '2020-01-01' and FECHA <='2020-12-31'")
df_2020.head(10)
```

```python
df_2020.shape
```

```python
#c.	Categorizar Condici√≥n
dfc = pd.get_dummies(df_2020, columns=["CONDICION"])
dfc.head(10)
```

```python
#d.	Categorizar estado
dfc = pd.get_dummies(df_2020, columns=["ESTADO"])
dfc.head(10)
```

```python
#e.	Edad, todos los datos num√©ricos
df_2020 = df_2020.drop(df_2020[df_2020['EDAD'] == "SIN INFORMACION"].index)
df_2020.head(10)
```

```python
df_2020.shape
```

```python
#f.	Categorizar sexo
dfc = pd.get_dummies(df_2020, columns=["SEXO"])
dfc.head(10)
```

```python
# g.    El orden de la data al final, debe ser:
#   i.  Fecha
#   ii. Condici√≥n
#   iii.    Edad
#   iv. Sexo
#   v.  Estado

df_2020 = df_2020[['FECHA','CONDICION','EDAD','SEXO','ESTADO']]
df_2020.head(10)
```
