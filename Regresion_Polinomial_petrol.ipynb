```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('petrol_consumption.csv')
```

```python
x = data.iloc[:,1:2].values
y = data.iloc[:, 2].values
```

```python
#grafica
plt.scatter(x,y,)
```

```python
#aplicando regresion lineal
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(x,y)
y_pred = lin.predict(x)
```

```python
#graficando
plt.scatter(x,y)
plt.plot(x,y_pred, color = 'red', linewidth = 5)
plt.title('Regresion lineal')
plt.xlabel('impuesto')
plt.ylabel('renta')
plt.show()
```

```python
#aplicando regresion polinomial
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=4)
x_poly = poly.fit_transform(x)
poly.fit(x_poly, y)
```

```python
lin2 = LinearRegression()
lin2.fit(x_poly, y)
y_pred2 = lin2.predict(x_poly)
```

```python
#graficando
plt.scatter(x,y)
plt.plot(x,y_pred2, color = 'red', linewidth = 5)
plt.title('Regresion lineal')
plt.xlabel('impuesto')
plt.ylabel('renta')
plt.show()
```
