```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values
```

```python
#dendogram
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title('dendogram customers')
plt.xlabel('registros de clientes')
plt.ylabel('Distancias')
dend = shc.dendrogram(shc.linkage(X, method='ward'))
```

```python
#aplicando algoritmo
from sklearn.cluster import  AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=5)
y_hc = cluster.fit_predict(X)
```

```python
#grafica
plt.figure(figsize=(15,10))
plt.scatter(X[y_hc==0,0], X[y_hc==0,1], c='red',label = "Cluster 1")
plt.scatter(X[y_hc==1,0], X[y_hc==1,1], c='blue',label = "Cluster 2")
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], c='green',label = "Cluster 3")
plt.scatter(X[y_hc==3,0], X[y_hc==3,1], c='brown',label = "Cluster 4")
plt.scatter(X[y_hc==4,0], X[y_hc==4,1], c='magenta',label = "Cluster 5")
plt.title("cluster jerarquico aglomerante clientes")
plt.xlabel('Nivel escolaridad')
plt.ylabel('Nivel de ingresos')
plt.legend()
plt.show()
```
